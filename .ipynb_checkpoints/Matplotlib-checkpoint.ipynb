## Scatter

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([8,2,3,6,8,9])
y = np.array([7,2,6,5,9,1])
plt.scatter(x, y)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([8,2,3,6,8,9])
y = np.array([7,2,6,5,9,1])
plt.scatter(x, y, c = "red")
x = np.array([4,7,8,2,8,6])
y = np.array([8,5,2,3,4,1])
plt.scatter(x, y, c = "blue")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([8,2,3,6,8,9])
y = np.array([7,2,6,5,9,1])
color = ["blue", "red", "yellow", "green", "purple", "black"]
plt.scatter(x, y, c = color)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array([8,2,3,6,8,9])
y = np.array([7,2,6,5,9,1])
color = np.array([0,20,40,60,80,100])
plt.scatter(x, y, c = color, cmap = 'viridis', s = 50, alpha = 0.3)
plt.colorbar()
```

## Bar Chart

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["Apple", "Orange", "Guava", "Stawberry"])
y = np.array([57,36,49,89])
plt.bar(x, y, color = "red")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["Apple", "Orange", "Guava", "Stawberry"])
y = np.array([57,36,49,89])
a = np.array([88,12,54,63])
b = np.array(["Car", "Bike", "Scooty", "Auto"])
plt.bar(x, y, color = "red")
plt.bar(b, a)
# Merges the two data
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["Apple", "Orange", "Guava", "Stawberry"])
y = np.array([57,36,49,89])
b = np.array(["Car", "Bike", "Scooty", "Auto"])
plt.bar(x, y, color = "red")
plt.xticks(x, b)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["Apple", "Orange", "Guava", "Stawberry"])
y = np.array([57,36,49,89])
# plt.bar(x, y, color = "green", width = 0.2)
plt.barh(x, y, color = "green", height = 0.2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.array(["Apple", "Orange", "Guava", "Stawberry"])
help = np.array([5,10,15,20])
y = np.array([57,36,49,89])
z = np.array([45,89,63,21])
plt.bar(help+1, y, color = "red")
plt.bar(help-1, z, color = "blue")
plt.xticks(help, x) 
```

## Pie Chart

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80,26,108,64,99])
mylabels = np.array(["Apple", "Guava", "Cherry", "Banana", "Pineapple"])
myexplode = np.array([0.2,0,0,0.2,0])
mycolors = np.array(["red", "green","maroon" ,"yellow", "orange"])
plt.pie(x, labels = mylabels, shadow = True, explode = myexplode, startangle = 270, autopct = '%0.2f%%', colors = mycolors)
# plt.legend(title = "Fruiut Stock in Kilograms")
plt.show()
```
